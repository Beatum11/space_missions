# Space Missions Analysys

```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv('space_missions.csv', encoding='latin1')
df.head()
```

```python
# Let's update the Price column and fill it with 0s where the values are missing.
```

```python
df['Price'] = df['Price'].fillna(0)
df.head()
```

## Question. Does success of a mission depend on a specific company?

```python
# To understand this, we need to do the following:
#    - Calculate the total number of successful missions for each company
#    - Compare the companies based on the total number of successful missions
```

```python
success_mis_df = df[df['MissionStatus'] == 'Success']
success_mis_df = success_mis_df.groupby('Company')['MissionStatus'].count().reset_index(name='TotalSuccess')
success_mis_df = success_mis_df.sort_values(by='TotalSuccess', ascending=False)
success_mis_df.head()
```

```python
# Now let's create a bar plot to visualize this, and we'll take the first 20 rows to avoid
# overloading the plot.
```

```python
first_20_comp = success_mis_df.head(20)
fig, ax = plt.subplots(1, 1, figsize=(12, 6))
ax.bar(first_20_comp['Company'], first_20_comp['TotalSuccess'])

ax.set_xlabel('Company')
ax.set_ylabel('Total Success')
ax.set_title('Top 20 Companies by Successful Space Missions')

ax.set_xticks(range(len(first_20_comp['Company'])))
ax.set_xticklabels(first_20_comp['Company'], rotation=45, ha='right')

plt.show()
```

```python
# During this analysis, it became interesting to compare missions throughout the years they were conducted. 
# I decided to take the first 8 companies and determine which of them are no longer active 
# and which have increased their number of missions.
```

```python
def missions_by_year(df: pd.DataFrame, company_name: str):
    temp_df = df[(df['Company'] == company_name) & (df['MissionStatus'] == 'Success')].copy()
    temp_df['Date'] = pd.to_datetime(temp_df['Date'])
    temp_df['Year'] = temp_df['Date'].dt.year
    temp_df = temp_df.groupby('Year')['MissionStatus'].count().reset_index(name='SuccessByYear')
    return temp_df
```

```python
# Let's compare first 4 companies
```

```python
companies = ['RVSN USSR','CASC', 'Arianespace', 'General Dynamics']

company_data = {}

for company in companies:
    company_data[company] = missions_by_year(df, company)

fig, axes = plt.subplots(2, 2, figsize=(14, 10))

# Use flatten() to convert an array into a one-dimensional array.
for ax, company in zip(axes.flatten(), companies):
    ax.plot(company_data[company]['Year'], company_data[company]['SuccessByYear'])
    ax.set_xlabel('Year of the missions')
    ax.set_ylabel('Amount of successful missions')
    ax.set_title(f'{company} successful space missions by year')

fig.tight_layout()
plt.show()
```

```python
# Second 4 companies
```

```python
companies = ['VKS RF', 'NASA', 'SpaceX', 'ULA']

company_data = {}

for company in companies:
    company_data[company] = missions_by_year(df, company)

fig, axes = plt.subplots(2, 2, figsize=(14, 10))

# Use faltten() to convert an array of axes into one-dimensional array.
for ax, company in zip(axes.flatten(), companies):
    ax.plot(company_data[company]['Year'], company_data[company]['SuccessByYear'])
    ax.set_xlabel('Year of the missions')
    ax.set_ylabel('Amount of successful missions')
    ax.set_title(f'{company} successful space missions by year')

fig.tight_layout()
plt.show()
```

```python
# What is clearly evident is that, although the RVSN USSR achieved the highest number of successful space missions, they no longer do so, at least because the USSR no longer exists.

# Secondly, CASC and SpaceX have demonstrated remarkable growth in the number of their missions. This trend is likely to continue.

# Therefore, it is only a matter of time before one of these companies surpasses the "record" set by the RVSN.
```

## Question. Has the number of successful launches changed over time?

```python
yearly_launches = df.copy()
yearly_launches['Date'] = pd.to_datetime(yearly_launches['Date'])
yearly_launches['Date'] = yearly_launches['Date'].dt.year

year_and_status_df = pd.DataFrame(yearly_launches[['Date', 'MissionStatus']])
status_by_year = year_and_status_df.groupby(['Date', 'MissionStatus']).size().unstack(fill_value=0)
status_by_year.head()
```

```python
colors_dict = {
    'Failure': 'red',
    'Partial Failure': 'orange',
    'Prelaunch Failure': 'blue',
    'Success': 'green'
}

status_by_year.plot(kind='bar', stacked=True, figsize=(14, 6), color=colors_dict)
plt.xlabel('Year')
plt.ylabel('Number of Missions')
plt.title('Number of Successful and Unsuccessful Missions Over Time')
plt.legend(title='Mission Status')
plt.show()
```

```python
# The number of successful launches has indeed changed over time, with notable peaks during the Cold War, 
# a period of stagnation following it, and a recent surge in activity. 
# These trends underscore the evolving nature of space exploration, influenced by geopolitical factors, 
# technological advancements, and the entry of private enterprises into the space industry.
```
